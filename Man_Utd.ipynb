```r
calculate_team_stats <- function(team_data) {
  data_summary <- team_data %>%
    summarise(
      Total_Points = sum(pts),
      Expected_Points = sum(xpts),
      Average_Score_Goals = mean(scored),
      Average_Expected_Goals = mean(xG),
      Average_Conceded_Goals= mean(missed),
      Average_Expected_Goals_Conceded = mean(xGA)
    )
  data_summary$Goal_Difference <- data_summary$Average_Score_Goals - data_summary$Average_Conceded_Goals
  data_summary$Expected_Goal_Difference <- data_summary$Average_Expected_Goals - data_summary$Average_Expected_Goals_Conceded
  data_summary$Chance_Conversion_Rate <- data_summary$Average_Score / data_summary$Average_Expected_Goals
  
  return(data_summary)
}
```

```r


teams <- c("Liverpool", "Man City", "Man Utd", "Chelsea")
team_stats <- lapply(teams, function(team) {
  team_data <- filter(Details, teamId == team)
  team_stats <- calculate_team_stats(team_data)
  team_stats$Team <- team 
  return(team_stats)
})
```

```r

team_stats_df <- bind_rows(team_stats)


team_stats_df <- team_stats_df %>%
  arrange(desc(Total_Points)) %>%
  select(Team, everything())
```

## Offense
As I mentioned earlier,points are awarded as follows: three points for a win, one point for a draw, and none for a loss.Therefore, as long as a team can score more goals than the opponent in every single game,regardless of defensive weaknesses, will win the title at the end of the season. So offensive performance is vital for any team aiming for a top position in the standings.

```r
library(ggplot2)
library(dplyr)
library(tidyr)

long_data <- team_stats_df %>%
  pivot_longer(cols = c(Average_Expected_Goals, Average_Score_Goals), names_to = "Metric", values_to = "Value")


bar_colors <- c("Average_Expected_Goals" = "blue", "Average_Score_Goals" = "orange")


Offensive_plot <- ggplot(long_data, aes(x = Team, y = Value, fill = Metric)) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = bar_colors) +
  theme_minimal(base_size = 14) +
  theme(
    plot.background = element_rect(fill = "white"),
    panel.background = element_rect(fill = "white"),
    plot.title = element_text(hjust = 0.5), 
    legend.text = element_text(size = 10)   
  ) +
  labs(x = "", y = "", title = "Offensive Performance Metrics", fill = "Metric") +
  geom_text(aes(label = round(Value, 2)), position = position_dodge(width = 0.9), vjust = -0.25)


ggsave("barplot.png", plot = Offensive_plot, bg = "white", width = 10, height = 6)

```

***Average Expected Goals***:It tells us how many goals a team likely to score based on their chances created in a game. A high xG can indicate an aggressive style of play, suggesting that the team frequently pushes forward, creates significant scoring opportunities, and, in essence, dominates the game's offensive flow. It's a metric that summarize the quality of the attacking system, the team's assertiveness in the final third, and their ability to impose their gameplay on the opposition and the control of the game. In short, xG is a comprehensive indicator of a team's offensive performance in a match.


***Average Scored Goals:*** It measures the actual number of goals a team scores on average each game. Unlike xG, which predicts scoring chances, this shows how many goals a team really puts on the scoreboard. High numbers here mean the team is not just getting chances; they're finishing them by scoring goals. It's a straightforward way to see how effective a team's offense is in real matches.

# Defense

Defense is just as important as offense in soccer. While teams like Manchester City, considered the best in offense, have an average expected goal rate of 2.61, it's unrealistic to expect four or five goals in every single game. Because goals can be hard to come by; not all chances lead to a goal, and luck plays its part too. 

Recognizing this, teams  particularly those battling to avoid relegation or in tightly contested matches where neither side wants to drop points  tend to emphasize defense. Their strategy is to maintain a solid back line to ensure at least a draw, choosing to play it safe rather than risk a loss in a high-scoring affair. Ultimately, a reliable defense is fundamental for any teams success.

```r
library(ggplot2)
library(dplyr)
library(tidyr)

long_data <- team_stats_df %>%
  pivot_longer(cols = c(Average_Expected_Goals_Conceded, Average_Conceded_Goals), names_to = "Metric", values_to = "Value")

bar_colors <- c("Average_Expected_Goals_Conceded" = "blue", "Average_Conceded_Goals" = "orange")


Defense_plot <- ggplot(long_data, aes(x = Team, y = Value, fill = Metric)) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = bar_colors) +
  theme_minimal(base_size = 14) +
  theme(
    plot.background = element_rect(fill = "white"),
    panel.background = element_rect(fill = "white"),
    plot.title = element_text(hjust = 0.5), 
    legend.text = element_text(size = 10)   
  ) +
  labs(x = "", y = "", title = "Defensive Performance Metrics", fill = "Metric") +
  geom_text(aes(label = round(Value, 2)), position = position_dodge(width = 0.9), vjust = -0.25)


ggsave("barplot.png", plot = Offensive_plot, bg = "white", width = 10, height = 6)

```

***Average Expected Goals Conceded:*** This stat predicts the number of goals a team might allow, taking into account the quality of their opponents' shots. It's a way to measure both how many and how good the chances are that they're giving up. A higher number means the team is likely giving their opponents opportunities that could easily turn into goals, suggesting potential weaknesses in defense.

***Average Conceded Goals:*** This is the actual average number of goals a team has given up per game. It tells us how well the team's defense is holding up in real-world scenarios. A low number here means the defense is strong and not many goals are being conceded, while a high number could be a sign that the team's defense is struggling to keep the ball out of their net.

# Conversion

Conversion in soccer is about turning created chances into actual goals, because goals ultimately decide the winner of the game. It's not just about how many opportunities a team makes; it's crucial they put the ball in the net. Sometimes, a team might dominate a match, controlling the play and creating numerous chances, has a very high expected goals, but it's the team that converts even a single opportunity to goal that can win the final wistle. This is the unpredictability of soccerdominance doesn't always lead to a win; it's the conversion of chances that counts. One shining moment, one swift strike can turn a game on its head, making conversion a key factor in any team's success.

```r

par(bg = "white")


barplot(team_stats_df$Chance_Conversion_Rate,
        main = "Goal Conversion Rate for Each Team",
        ylab = "Conversion Rate",
        names.arg = team_stats_df$Team,
        col = "blue",
        horiz = FALSE)


par(bg = "transparent")
```

From the barplot, we observe the Goal Conversion Rate for Liverpool, Man City, Chelsea, and Man Utd. Liverpool, which clinched the title, has a remarkable conversion rate that exceeds 1. This impressive figure implies that Liverpool has been exceptionally efficient at scoring; they have not only converted the chances created but also been adept at scoring from less probable situations. This could be due to striking wonder goals from distance or capitalizing on opponents' mistakes to score.

Manchester City, Chelsea, and Manchester United follow Liverpool with conversion rates below 1, which indicates that while they are converting chances into goals, they haven't done so with the same exceptional efficiency as Liverpool. Manchester United, in particular, has the lowest conversion rate among the four, suggesting that there may be areas for improvement in turning their chances into goals.
